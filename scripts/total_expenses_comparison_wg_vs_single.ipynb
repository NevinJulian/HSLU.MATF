```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Monatliche Kosten (CHF)
miete_single = 1500
nebenkosten_single = 350
internet_single = 60

miete_wg = 1000
nebenkosten_gesamt = 350
internet_gesamt = 60

# WG-Größen zur Analyse
wg_groessen = [2, 3, 4, 5]

# Zeitachse: Monate über 10 Jahre
monate = np.arange(0, 10 * 12 + 1)
jahre = monate / 12
```

```python
# Berechnung der kumulierten Kosten
kosten_single = (miete_single + nebenkosten_single + internet_single) * monate
kosten_wg_varianten = {}
kostendifferenzen = {}

for n in wg_groessen:
    kosten_pro_person = (miete_wg + nebenkosten_gesamt / n + internet_gesamt / n) * monate
    kosten_wg_varianten[n] = kosten_pro_person
    kostendifferenzen[n] = kosten_single - kosten_pro_person
```

```python
# Plot der kumulierten Kosten
plt.figure(figsize=(12, 6))
plt.plot(jahre, kosten_single, label='Singlewohnung', linewidth=2)

for n in wg_groessen:
    plt.plot(jahre, kosten_wg_varianten[n], label=f'WG mit {n} Personen', linestyle='--')

plt.title('Kumulierte Wohnkosten: Singlewohnung vs. WG-Größen')
plt.xlabel('Jahre')
plt.ylabel('Kumulierte Kosten (CHF)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# %%
# Plot der Ersparnis gegenüber Singlewohnung
# ----------------------------

plt.figure(figsize=(12, 6))
for n in wg_groessen:
    plt.plot(jahre, kostendifferenzen[n], label=f'Single - WG ({n} Personen)')

plt.title('Kostendifferenz: Singlewohnung vs. WG-Größen')
plt.xlabel('Jahre')
plt.ylabel('Ersparnis (CHF)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
